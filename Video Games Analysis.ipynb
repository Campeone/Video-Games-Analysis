# Exploratory Data Analysis 


## Platform

```python
vgs.Platform.unique()
```

```python
# platform value counts 
vgs.Platform.value_counts()
```

```python
plt.figure(figsize = (14, 5))
vgs.Platform.value_counts().plot(kind = 'bar', title = 'Video Games Platform', xlabel = 'Platform', ylabel = 'Frewuency')
```

### Most succesful Video Games Platform

```python
# Top 10 succesful Video Games platform
Top_10_Platform = vgs.groupby('Platform')['Global_Sales'].sum().sort_values(ascending = False).head(10)
Top_10_Platform
```

```python
plt.figure(figsize = (12, 5))
Top_10_Platform.plot(kind = 'bar', title = 'Video Games Platform', xlabel = 'Platform', ylabel = 'Total Global Sales')
```

### Video Games with the Highest Global Sales for each Platform

```python
vgs.loc[vgs.groupby('Platform').Global_Sales.idxmax()].sort_values('Global_Sales', ascending = False)
```

## Genre

```python
vgs.Genre.unique()
```

```python
# value counts 
vgs.Genre.value_counts()
```

```python
# bar chart
plt.figure(figsize = (14, 5))
vgs.Genre.value_counts().plot(kind = 'bar', title = 'Video Games Genre', xlabel = 'Genre', ylabel = 'Frequency')
```

### Most Profitable Video Games Genre

```python
top_genre_by_sales = vgs.groupby('Genre').Global_Sales.sum().sort_values(ascending = False) 
top_genre_by_sales
```

```python
plt.figure(figsize = (12, 5))
top_genre_by_sales.plot(kind = 'bar', title = 'Genre by Global Sales', xlabel = 'Genre', ylabel = 'Total Global Sales')
```

### Video Games with the Highest Global Sales for each Genre

```python
vgs.loc[vgs.groupby('Genre').Global_Sales.idxmax()].sort_values('Global_Sales', ascending = False)
```

## Year

```python
vgs.Year.unique()
```

```python
# Year value counts
vgs_year_count = vgs.Year.value_counts().sort_index() 
vgs_year_count
```

```python
plt.figure(figsize = (15, 5))
vgs_year_count.plot(title = 'Video Games Yearly Count', xlabel = 'Year', ylabel = 'Frequency')
```

### Yearly Global Sales Trend

```python
yearly_global_sales = vgs.groupby('Year').Global_Sales.sum()
yearly_global_sales
```

```python
# yearly global sales trend
plt.figure(figsize = (12, 5))
yearly_global_sales.plot(kind = 'line', title = 'Yearly Global Sales Trend', ylabel = 'Global Sales')
```

### Video Games with the Top Global Sales per Year

```python
vgs.loc[vgs.groupby('Year').Global_Sales.idxmax()]
```
